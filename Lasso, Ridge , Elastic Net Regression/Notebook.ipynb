# Lasso Regression 


Lasso Regression is done to improve the generalizability of the model. This is done by tweaking the slope of the best fit line. Maybe the model does not perform much well in the training because now the line does not pass exactly to the data points but it will give pretty good results in testing. The slope is changed or the line is titled a bit by making use of the penalty term called alpha which is a hyperparameter.
