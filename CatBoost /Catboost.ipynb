# Catboost
- Catboost can take in categorical, Numerical, and text features but excels with categorical data
- Created in 2017 by Yandex Researchers 
- Some use cases involve reccomendation systems, self-driving cars, personal assistants, weather prediction, etc.
Why it performs better than other boosting algos:

- Excellent quality without the need for parameter tuning 
    - Reduction in parameter tuning because default produces great results
- Support for categorical features 
    - Can use non numerical variables to avoid the heavy task of one hot encoding/ mapping variables 
- GPU version that is both fast and scalable ]
    - Done using fast GPU implentation of the gradient boost algo. 
- Excellent Accuracy 
    - Novel gradient boosting schema can be used to avoid overfitting the model.
- Rapid Predictions 
     - Can be applied at a rapid pace with quick results
