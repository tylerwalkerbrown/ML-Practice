# Linear Regression
- Linear Regression is a ML model that takes in continous variables such as weight, temp, etc. This model is built around one independent variable and a dependent variable. The independent being the input data and dependent being the target output. Taking in this equation to the graph you can predict what a variables output is going to be (most likely).







The model uses what is called the ordinary least squares method that fits the line to the dat ausing the error. The goal of this is to get the lowest possible error to return the best results with the lowest residuals. 


```python
#Iris Dataset 
from sklearn.datasets import load_boston

dataset = pd.read_csv('https://gitlab.com/AnalyticsIndiaMagazine/practicedatasets/-/raw/main/bootcamp/simple_linear_regression/Salary_Data.csv')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score 
```

# Training and Test Patterns

```python
#Grabbing the taget and inputs
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values


# Checking the shape of input and output features
print('Shape of the input features:', x.shape)
print('Shape of the output features:', y.shape)
```

```python
# Defining the training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
# Checking the shape of the training and test sets
print('Shape of the test data:', y_train.shape)
```

# Model Creation and Fit 

```python
#Creating model object 
regres = LinearRegression()
```

```python
regres.fit(x,y)
```

```python
print('The regression coefficient:', regres.coef_[0])
```

```python
print('Intercept:' , regres.intercept_)
```

```python
print('The regression line is: y = ', regres.intercept_, '+', regres.coef_[0], '*X]')
```

```python
y_pred = regres.predict(X_test)
```

```python
#Predicted Values
y_pred
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regres.predict(X_train), color = 'blue')
plt.title('Salary vs Experience')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

# Eval Metrics

```python
# Mean Squared Error (MSE)
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test, y_pred)
print('Mean Squared Error is:', MSE)
```

```python
# Root Mean Squared Error (RMSE)
import math
RMSE = math.sqrt(MSE)
print('Root Mean Squared Error is:', RMSE)
```

```python
# R-Squared (Strong Model)
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print('R-Squared is:', r2)
```
